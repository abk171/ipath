```python
import json
import pandas as pd
```

```python
gnomad_path = './output/gnomad_data.json'
clinvar_path = './output/clinvar_data.json'
gnomad_more_path = './output/gnomad_data_more.json'
```

```python
with open(gnomad_path, 'r') as file:
    gnomad_data = json.load(file)
    
with open(clinvar_path, 'r') as file:
    clinvar_data = json.load(file)
    
with open(gnomad_more_path, 'r') as file:
    gnomad_more_data = json.load(file)

gnomad_data.extend(gnomad_more_data)
```

```python
print(json.dumps(gnomad_data[0], indent = 4))
```

```python
print(json.dumps(clinvar_data[0], indent = 4))
```

```python
gnomad_df = pd.DataFrame()
for i, data in enumerate(gnomad_data):
    try:
        variants = data['gene']['variants']
        variants_data = pd.json_normalize(variants)
        variants_data['gene_id'] = data['gene']['gene_id']
        gnomad_df = gnomad_df.append(variants_data, ignore_index=True)
    except Exception as e:
        pass
       
```

```python
clinvar_df = pd.DataFrame()
for i, data in enumerate(clinvar_data):
    try:
        variants = data['gene']['clinvar_variants']
        clinvar_df = clinvar_df.append(pd.json_normalize(variants), ignore_index=True)
    except Exception as e:
        pass
```

```python
gnomad_benign_df = gnomad_df[gnomad_df['transcript_consequence.polyphen_prediction'].isin(['benign', 'unknown', None])]
```

```python
gnomad_benign_notclinvar_df = gnomad_benign_df[~gnomad_benign_df['variant_id'].isin(clinvar_df['variant_id'])]
```

```python
final_gnomad_benign = gnomad_benign_notclinvar_df.groupby('gene_id').head(98)
```
